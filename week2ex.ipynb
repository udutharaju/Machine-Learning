```python
##12  Write a function to find min and max values of any column
def find_min_max(df, column_name):
    min_value = df[column_name].min()
    max_value = df[column_name].max()
    return min_value, max_value
```

```python
#13. Call the above function to find min max of horsepower, length, fuel efficiency
# Call the function to find min and max for each column
hp_min, hp_max = find_min_max(df, 'Horsepower')
length_min, length_max = find_min_max(df, 'Length')
fuel_efficiency_min, fuel_efficiency_max = find_min_max(df, 'Fuel efficiency')

# Print the results
print("\nHorsepower - Min:", hp_min, "Max:", hp_max)
print("Length - Min:", length_min, "Max:", length_max)
print("Fuel Efficiency - Min:", fuel_efficiency_min, "Max:", fuel_efficiency_max)
```
